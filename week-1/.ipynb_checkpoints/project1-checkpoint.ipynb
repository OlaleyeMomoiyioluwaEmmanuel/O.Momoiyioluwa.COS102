```python
# Financial Calculations: Simple Interest, Compound Interest, Annuity Plan

def simple_interest(P, R, T):
    A = P * (1 + (R/100) * T)
    return A

def compound_interest(P, R, n, t):
    A = P * (1 + R/n) ** (n*t)
    return A

def annuity_plan(PMT, R, n, t):
    A = PMT * (((1 + R/n) ** (n*t) - 1) / (R/n))
    return A

# Example usage
P = 1000   # Principal
R = 5      # Interest rate (5%)
T = 2      # Time in years
n = 4      # Compounding periods per year
PMT = 200  # Payment per period

print("Simple Interest Amount:", simple_interest(P, R, T))
print("Compound Interest Amount:", compound_interest(P, R/100, n, T))
print("Annuity Plan Amount:", annuity_plan(PMT, R/100, n, T))
```
