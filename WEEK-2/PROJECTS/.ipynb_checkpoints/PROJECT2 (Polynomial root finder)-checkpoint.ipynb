## import numpy as np
import math

print("Polynomial Root Finder")
print("1. Quadratic Equation (Ax^2 + Bx + C = 0)")
print("2. Cubic Equation (Ax^3 + Bx^2 + Cx + D = 0)")
print("3. Quartic Equation (Ax^4 + Bx^3 + Cx^2 + Dx + E = 0)")

choice = int(input("Choose the operation (1/2/3): "))

if choice == 1:
    print("\nQuadratic Equation")
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))

    discriminant = b**2 - 4*a*c

    if discriminant >= 0:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
    else:
        root1 = (-b + complex(0, math.sqrt(-discriminant))) / (2*a)
        root2 = (-b - complex(0, math.sqrt(-discriminant))) / (2*a)

    print("Roots:", root1, root2)

elif choice == 2:
    print("\nCubic Equation")
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))

    coeff = [a, b, c, d]
    roots = np.roots(coeff)

    print("Roots:", roots)

elif choice == 3:
    print("\nQuartic Equation")
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))
    e = float(input("Enter E: "))

    coeff = [a, b, c, d, e]
    roots = np.roots(coeff)

    print("Roots:", roots)

else:
    print("Invalid choice")
