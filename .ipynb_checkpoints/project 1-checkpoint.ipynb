````python
START

total = 0

while True:
    scan_item()
    price = get_price()
    total = total + price
    
    more = input("More items? (yes/no): ")
    
    if more == "no":
        break

print("Total amount:", total)

payment = float(input("Enter payment amount: "))

if payment >= total:
    change = payment - total
    print("Give change:", change)
    print("Print receipt")
else:
    print("Ask for balance")

END
````


```mermaid
flowchart TD

A([Start]) --> B[Scan Item]
B --> C[Add Price to Total]
C --> D{More Items?}

D -- Yes --> B
D -- No --> E[Display Total]

E --> F[Receive Payment]
F --> G{Payment ≥ Total?}

G -- Yes --> H[Give Change & Print Receipt]
G -- No --> I[Ask for balance]

H --> J([End])
I --> J
