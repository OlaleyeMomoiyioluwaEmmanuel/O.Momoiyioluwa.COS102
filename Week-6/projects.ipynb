```python
import tkinter as tk
from tkinter import messagebox
import random

# the raw data
employees_list = [
    "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali",
    "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James",
    "Michell Taiwo", "Abraham Jones", "Nicole Anide",
    "Kosi Korso", "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
]

tasks_list = [
    "Loading", "Transporting", "Reviewing Orders",
    "Customer Service", "Delivering Items"
]

# the Classes
class Employee:
    def __init__(self, name):
        self.name = name

    def check_employee(self):
        return self.name in employees_list

    def take_attendance(self):
        return f"Attendance taken for {self.name}"

    def assign_task(self):
        task = random.choice(tasks_list)
        return f"Task assigned: {task}"

    def refuse_access(self):
        return "Access Denied! You are not one of us."


# this is for theapp itself
class App:
    def __init__(self, root):
        self.root = root
        self.root.title("Employee Verification System")
        self.root.geometry("400x300")

        self.label = tk.Label(root, text="Enter Employee Name:")
        self.label.pack(pady=10)

        self.entry = tk.Entry(root, width=30)
        self.entry.pack(pady=10)

        self.button = tk.Button(root, text="Submit", command=self.process)
        self.button.pack(pady=10)

        self.result = tk.Label(root, text="", wraplength=300)
        self.result.pack(pady=20)

    def process(self):
        name = self.entry.get().strip()
        emp = Employee(name)

        if emp.check_employee():
            attendance = emp.take_attendance()
            task = emp.assign_task()
            self.result.config(text=f"{attendance}\n{task}")
        else:
            self.result.config(text=emp.refuse_access())
            messagebox.showerror("Error", "Access Denied!")


# Run App
if __name__ == "__main__":
    root = tk.Tk()
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

# OOP Class
class Delivery:
    def __init__(self, location, weight):
        self.location = location.lower()
        self.weight = weight

    def calculate_fee(self):
        if self.location == "pau":
            if self.weight >= 10:
                return 2000
            else:
                return 1500

        elif self.location == "epe":
            if self.weight >= 10:
                return 5000
            else:
                return 4000
        else:
            return None


# GUI Class
class App:
    def __init__(self, root):
        self.root = root
        self.root.title("Delivery Price Calculator")
        self.root.geometry("400x300")

        # Location input
        tk.Label(root, text="Enter Location (PAU or Epe):").pack(pady=10)
        self.location_entry = tk.Entry(root)
        self.location_entry.pack()

        # Weight input
        tk.Label(root, text="Enter Package Weight (kg):").pack(pady=10)
        self.weight_entry = tk.Entry(root)
        self.weight_entry.pack()

        # Button
        tk.Button(root, text="Calculate Fee", command=self.process).pack(pady=15)

        # Result
        self.result_label = tk.Label(root, text="")
        self.result_label.pack(pady=10)

    def process(self):
        location = self.location_entry.get()
        
        try:
            weight = float(self.weight_entry.get())
        except ValueError:
            messagebox.showerror("Error", "Please enter a valid number for weight.")
            return

        delivery = Delivery(location, weight)
        fee = delivery.calculate_fee()

        if fee is None:
            messagebox.showerror("Error", "Invalid location! Use PAU or Epe.")
        else:
            self.result_label.config(text=f"Delivery Fee: ₦{fee}")


# Run App
if __name__ == "__main__":
    root = tk.Tk()
    app = App(root)
    root.mainloop()
```
